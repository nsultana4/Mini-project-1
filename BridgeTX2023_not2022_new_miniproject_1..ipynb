```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt

# Function to convert DMS to Decimal Degrees and adjust based on WGS 84 for Texas
def dms_to_dd(dms_value, is_latitude=True):
    degrees = int(dms_value / 1000000)
    minutes = int((dms_value - degrees * 1000000) / 10000)
    seconds = (dms_value - degrees * 1000000 - minutes * 10000) / 100
    
    decimal_degrees = degrees + (minutes / 60) + (seconds / 3600)
    
    # For latitudes, positive in the northern hemisphere
    if is_latitude:
        return decimal_degrees  # Texas is in the northern hemisphere, so positive latitude
    
    # For longitudes, negative in the western hemisphere
    else:
        return -decimal_degrees  # Texas is in the western hemisphere, so negative longitude

# Load the datasets
tx22 = pd.read_csv(r'E:\Users\Mini Project -1\OneDrive\Desktop\GIS Data\tx22.csv')
tx23 = pd.read_csv(r'E:\Users\Mini Project -1\OneDrive\Desktop\GIS Data\tx23.csv')

# Find bridges that are in 2023 but not in 2022
bridges_2023_not_2022 = tx23[~tx23['STRUCTURE_NUMBER_008'].isin(tx22['STRUCTURE_NUMBER_008'])]

# Convert latitude and longitude from DMS to Decimal Degrees with proper sign
bridges_2023_not_2022['LAT_016_dd'] = bridges_2023_not_2022['LAT_016'].apply(lambda x: dms_to_dd(x, is_latitude=True))
bridges_2023_not_2022['LONG_017_dd'] = bridges_2023_not_2022['LONG_017'].apply(lambda x: dms_to_dd(x, is_latitude=False))

# Create a GeoDataFrame with the converted coordinates
geometry = [Point(xy) for xy in zip(bridges_2023_not_2022['LONG_017_dd'], bridges_2023_not_2022['LAT_016_dd'])]
gdf = gpd.GeoDataFrame(bridges_2023_not_2022, geometry=geometry)

# Set the CRS to WGS84 (EPSG:4326)
gdf.set_crs(epsg=4326, inplace=True)

# Plotting the map
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, color='red', marker='o', markersize=1, label='Bridges 2023 not 2022')

# Customize the map
ax.set_title('Bridges Surveyed in 2023 but Not in 2022', fontsize=15)
ax.set_xlabel('Longitude', fontsize=12)
ax.set_ylabel('Latitude', fontsize=12)
plt.legend()
plt.grid(True)

# Display the plot
plt.show()

# Save to shapefile
gdf.to_file(r'E:\Mini Project -1\OneDrive\Desktop\bridges_2023_not_2022.shp')

print("Shapefile has been saved as 'bridges_2023_not_2022.shp' with WGS84 CRS.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

# Load the datasets
tx22 = pd.read_csv(r'C:\Users\asidd\OneDrive\Desktop\GIS Data\tx22.csv', low_memory=False)
tx23 = pd.read_csv(r'C:\Users\asidd\OneDrive\Desktop\GIS Data\tx23.csv', low_memory=False)

# Find bridges that are in 2023 but not in 2022
bridges_2023_not_2022 = tx23[~tx23['STRUCTURE_NUMBER_008'].isin(tx22['STRUCTURE_NUMBER_008'])]

# Convert string ratings to numeric, handling non-numeric values
bridges_2023_not_2022['DECK_COND_058'] = pd.to_numeric(bridges_2023_not_2022['DECK_COND_058'], errors='coerce')
bridges_2023_not_2022['SUPERSTRUCTURE_COND_059'] = pd.to_numeric(bridges_2023_not_2022['SUPERSTRUCTURE_COND_059'], errors='coerce')
bridges_2023_not_2022['SUBSTRUCTURE_COND_060'] = pd.to_numeric(bridges_2023_not_2022['SUBSTRUCTURE_COND_060'], errors='coerce')

# Preprocess the ratings into binary classes (0 = Less than Satisfactory, 1 = Satisfactory or better)
def create_binary_rating(rating):
    if pd.isna(rating):
        return None  # or some other handling for NaNs, e.g., fill with a specific value or drop
    elif rating >= 5:
        return 1  # Satisfactory or better
    else:
        return 0  # Less than satisfactory

# Apply this function to the relevant columns using the correct names
bridges_2023_not_2022['deck_rating_binary'] = bridges_2023_not_2022['DECK_COND_058'].apply(create_binary_rating)
bridges_2023_not_2022['superstructure_rating_binary'] = bridges_2023_not_2022['SUPERSTRUCTURE_COND_059'].apply(create_binary_rating)
bridges_2023_not_2022['substructure_rating_binary'] = bridges_2023_not_2022['SUBSTRUCTURE_COND_060'].apply(create_binary_rating)

# Select features for prediction (customized based on available data)
features = ['AGE', 'MATERIAL', 'TRAFFIC', 'SPAN_LENGTH', 'LAT_016', 'LONG_017']
X = bridges_2023_not_2022[features]

# Handle missing values using SimpleImputer
imputer = SimpleImputer(strategy='mean')  # Replace missing numerical values with the mean
X[['AGE', 'SPAN_LENGTH', 'LAT_016', 'LONG_017']] = imputer.fit_transform(X[['AGE', 'SPAN_LENGTH', 'LAT_016', 'LONG_017']])

# Handle categorical features like MATERIAL and TRAFFIC (use OneHotEncoder for this)
encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
categorical_features = X[['MATERIAL', 'TRAFFIC']].fillna('Unknown')  # Fill missing categories with 'Unknown'

# One-hot encode the categorical features and merge them back into X
encoded_categorical_features = pd.DataFrame(encoder.fit_transform(categorical_features), 
                                            columns=encoder.get_feature_names_out(categorical_features.columns))
X = X.drop(['MATERIAL', 'TRAFFIC'], axis=1)
X = pd.concat([X.reset_index(drop=True), encoded_categorical_features.reset_index(drop=True)], axis=1)

# Predict ratings for Deck, Superstructure, and Substructure separately

# Deck Rating Prediction
y_deck = bridges_2023_not_2022['deck_rating_binary']
X_train_deck, X_test_deck, y_train_deck, y_test_deck = train_test_split(X, y_deck, test_size=0.3, random_state=42)

# Superstructure Rating Prediction
y_superstructure = bridges_2023_not_2022['superstructure_rating_binary']
X_train_superstructure, X_test_superstructure, y_train_superstructure, y_test_superstructure = train_test_split(X, y_superstructure, test_size=0.3, random_state=42)

# Substructure Rating Prediction
y_substructure = bridges_2023_not_2022['substructure_rating_binary']
X_train_substructure, X_test_substructure, y_train_substructure, y_test_substructure = train_test_split(X, y_substructure, test_size=0.3, random_state=42)

# Initialize RandomForestClassifier (or use another model of your choice)
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train model for Deck Rating
clf.fit(X_train_deck, y_train_deck)
y_pred_deck = clf.predict(X_test_deck)
print("Deck Rating Classification Report:")
print(classification_report(y_test_deck, y_pred_deck, zero_division=1))

# Train model for Superstructure Rating
clf.fit(X_train_superstructure, y_train_superstructure)
y_pred_superstructure = clf.predict(X_test_superstructure)
print("Superstructure Rating Classification Report:")
print(classification_report(y_test_superstructure, y_pred_superstructure, zero_division=1))

# Train model for Substructure Rating
clf.fit(X_train_substructure, y_train_substructure)
y_pred_substructure = clf.predict(X_test_substructure)
print("Substructure Rating Classification Report:")
print(classification_report(y_test_substructure, y_pred_substructure, zero_division=1))
# Make a copy to ensure you're not modifying the slice directly
bridges_2023_not_2022 = tx23[~tx23['STRUCTURE_NUMBER_008'].isin(tx22['STRUCTURE_NUMBER_008'])].copy()

# Conversion and assignment using .loc
bridges_2023_not_2022.loc[:, 'DECK_COND_058'] = pd.to_numeric(bridges_2023_not_2022['DECK_COND_058'], errors='coerce')
bridges_2023_not_2022.loc[:, 'SUPERSTRUCTURE_COND_059'] = pd.to_numeric(bridges_2023_not_2022['SUPERSTRUCTURE_COND_059'], errors='coerce')
bridges_2023_not_2022.loc[:, 'SUBSTRUCTURE_COND_060'] = pd.to_numeric(bridges_2023_not_2022['SUBSTRUCTURE_COND_060'], errors='coerce')

# Applying function using .loc
bridges_2023_not_2022.loc[:, 'deck_rating_binary'] = bridges_2023_not_2022['DECK_COND_058'].apply(create_binary_rating)
bridges_2023_not_2022.loc[:, 'superstructure_rating_binary'] = bridges_2023_not_2022['SUPERSTRUCTURE_COND_059'].apply(create_binary_rating)
bridges_2023_not_2022.loc[:, 'substructure_rating_binary'] = bridges_2023_not_2022['SUBSTRUCTURE_COND_060'].apply(create_binary_rating)
```
